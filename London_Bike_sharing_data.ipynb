```python
#Specifying the column names that I want to use
new_cols_dict={
    'timestamp':'time',
    'cnt':'count',
    't1':'temp_real_c',
    't2':'tem_feels_like_c',
    'hum':'humidity_percent',
    'wind speed':'wind_speed_kph',
    'weather_code':'weather',
    'is_holiday':'is_holiday',
    'is_weekend':'is_weekend',
    'season':'season'
}
```

```python
#Renaming the columns to the specified column names
bikes.rename(new_cols_dict, axis=1, inplace=True)
```

```python
#Changing the humidity values to percentage(i.e. a value between 0 and 1)
bikes.humidity_percent = bikes.humidity_percent / 100
```

```python
#Creating a season dictionary so that we can map the integers 0-3 to the actual written values
season_dict = {
    '0.0':'spring',
    '1.0':'summer',
    '2.0':'autum',
    '3.0':'winter'
 }
```

```python
#Creating a weather dictionary so that we cam map the integers to the actual written values
weather_dict = {
    '1.0':'clear',
    '2.0':'scattered clouds',
    '3.0':'broken clouds',
    '4.0':'cloudy',
    '7.0':'rain',
    '10.0':'rain with thunderstorm',
    '26.0':'snowfall'

}
```

```python
#Changing the seasons column data type to sring
bikes.season = bikes.season.astype('str')
```

```python
#Mapping the values 0-3 to the actual written seasons
bikes.season = bikes.season.map(season_dict)
```

```python
#Changing the weather column data type to sring
bikes.weather = bikes.weather.astype('str')
```

```python
#Mapping the value to the actual wrriten weathers
bikes.weather = bikes.weather.map(weather_dict)
```

```python
#Checking my dataframe to see if the mapping data has worked
bikes.head()
```

```python
#Writing the final dataframe to excel file that we will use in our tableau visualisation the fil will be the 'london_bikes.xlsx' and the sheet name is data
bikes.to_excel('london_bikes.xlsx', sheet_name='Data')
```
